```python
import pandas as pd

df = pd.read_csv(r"C:\Users\siven\Desktop\icici-nav.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')
df = df.set_index('date')

yearly = df['nav'].resample('Y').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last'
})

yearly
```

```python
df = yearly.copy()

df["change"] = df["close"] - df["open"]
df["change_percentage"] = df["change"] / df["open"] * 100


df["rolling_change_3"] = df["change_percentage"].rolling(window=3).mean()
df["rolling_change_5"] = df["change_percentage"].rolling(window=5).mean()
df["rolling_change_8"] = df["change_percentage"].rolling(window=8).mean()
df["rolling_change_10"] = df["change_percentage"].rolling(window=10).mean()



df = df.round(2)
df
```
